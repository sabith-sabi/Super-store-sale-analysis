## Product Analysis

```python
df.head(5)
```

```python
segment_counts = df['Category'].value_counts()
print("Segment counts:\n", segment_counts)
```

```python
# Group by "Category" and count unique "Subcategory" values within each group
subcategory_counts = df.groupby('Category')['Sub-Category'].nunique().reset_index()

# Display the result
subcategory_counts

```

```python

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(category_sales['Category'], category_sales['Sales'])
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the bar chart
plt.tight_layout()  # Ensures proper layout
plt.show()
```

```python
# Assuming your DataFrame is named 'df'
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().reset_index()
subcategory_sales = subcategory_sales.sort_values(by='Sales', ascending=True)

# Display the DataFrame
print(subcategory_sales)
```

```python

# Create a bar chart
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.bar(subcategory_sales['Sub-Category'], subcategory_sales['Sales'])
plt.xlabel('Sub-Category')
plt.ylabel('Sum of Sales')
plt.title('Sum of Sales by Subcategory')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better visibility

# Display the bar chart
plt.show()
```

## Sales

```python

# Assuming your DataFrame is named 'df'
df['Order Date'] = pd.to_datetime(df['Order Date'])  # Convert to datetime

# Extract the year from 'Order Date' and create a new 'Year' column
df['Year'] = df['Order Date'].dt.year

# Group by 'Year' and calculate the sum of sales
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# Display the result
print(yearly_sales)
```

```python

# Assuming you have the 'yearly_sales' DataFrame
plt.figure(figsize=(10, 6))
plt.bar(yearly_sales['Year'].astype(str), yearly_sales['Sales'])
plt.xlabel('Year')
plt.ylabel('Sum of Sales')
plt.title('Sum of Sales by Year')

# Display the bar chart
plt.show()
```

```python
# Assuming you have the 'yearly_sales' DataFrame
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales['Year'], yearly_sales['Sales'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Sum of Sales')
plt.title('Sum of Sales by Year')
plt.grid(True)

# Display the line graph
plt.show()
```

# Monthly Sales

```python
# Assuming your DataFrame is named 'df'
df['Order Date'] = pd.to_datetime(df['Order Date'])  # Convert to datetime

# Extract the year and month from 'Order Date'
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Filter the data for the year 2015
df_2015 = df[df['Year'] == 2015]

# Group by 'Year' and 'Month' and calculate the sum of sales
monthly_sales_2015 = df_2015.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Sort the result in ascending order of the month
monthly_sales_2015 = monthly_sales_2015.sort_values(by=['Year', 'Month'], ascending=[True, True])

# Display the result
monthly_sales_2015
```

```python
# Assuming you have the 'monthly_sales_2018' DataFrame
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales_2015['Month'], monthly_sales_2015['Sales'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend for 2015')
plt.xticks(monthly_sales_2015['Month'])
plt.grid(True)

# Display the line graph
plt.show()
```
